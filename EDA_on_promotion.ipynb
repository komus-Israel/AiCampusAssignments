```python
#import pandas as pd

import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt


#read into the file
df = pd.read_csv('train.csv')
```

```python
#sort the data via the promotion variable in descending order

df.sort_values(by = 'Promoted_or_Not', ascending = False)
```

```python
#to know the total numbers of promoted and not promoted


promotion_status_of_employees = df['Promoted_or_Not'].value_counts()

print('Employees not promoted = {} \nEmployees promoted = {}'.format(promotion_status_of_employees[0],promotion_status_of_employees[1]))
```

```python
#the data of workers that were promoted

df[df['Promoted_or_Not'] == 1]
```

```python
#the data of workers that were not promoted

df[df['Promoted_or_Not'] == 0]
```

```python
#removal of unnecessary variables

updated_data_frame = df.drop(['EmployeeNo','Division','Qualification','Channel_of_Recruitment','Year_of_recruitment','Gender','Year_of_birth','Previous_Award','State_Of_Origin','Foreign_schooled','Marital_Status','Past_Disciplinary_Action','Previous_IntraDepartmental_Movement'], axis = 1, inplace = False)
```

```python
#a function to cross tabulate the required variables to the promotion status


def cross_tab(feature):
    #if feature in df.columns:
    return pd.crosstab(updated_data_frame['Promoted_or_Not'], updated_data_frame[feature], margins = True)
    

```

```python
#print the tabulated features
for features in updated_data_frame.columns[:-1]:
    print(cross_tab(features)); print('-' * 100)
```

```python
#to visualize the data, input the feature_no corresponding to the feature_names as a parameter in the visualised_data()

def visualised_data(feature_no):
    feature_names = ['Trainings_Attended', 'Last_performance_score', 'Targets_met', 'Training_score_average', 'No_of_previous_employers']
    feature_numbers = [0,1,2,3,4]
    %config InlineBackend.figure_format = 'svg'
    for i,j in zip(feature_names, feature_numbers):
        if feature_no == j:
            return sn.countplot(x = i, hue = 'Promoted_or_Not', data = updated_data_frame)

```
